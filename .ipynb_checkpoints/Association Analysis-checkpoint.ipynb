```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

df = pd.read_excel('Online Retail.xlsx')
```

```python
df.head()
```

## Preprocessing

```python
# Remove leading and trailing whitespace from descriptions
df['Description'] = df['Description'].str.strip()

# Drop rows that don't have invoice numbers
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
```

```python
# Remove credit transactions (those with invoice numbers containing 'C')
df['InvoiceNo'] = df['InvoiceNo'].astype('str')
df = df[~df['InvoiceNo'].str.contains('C')]
```

Consolidate items into 1 transaction per row, with each product 1 hot encoded.

For sake of keeping the dataset small, we look at sales for France only

```python
basket = (df[df['Country'] == 'France']
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
```

```python
basket.head()
```

```python
# Convert units to 1 hot encoded values
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
```

```python
basket_sets = basket.applymap(encode_units)
```

```python
# Drop postage column since that charge is not one we wish to explor
basket_sets.drop('POSTAGE', inplace=True, axis=1)
```
